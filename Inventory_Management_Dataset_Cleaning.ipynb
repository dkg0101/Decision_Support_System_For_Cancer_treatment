```python
data6= pd.read_excel("Warehouse_Info_Inconsistency_v3.xlsx")
data6.head()
```

## Inspect Data Structure

```python
#Check the shape of data
print(data1.shape)
print(data2.shape)
print(data3.shape)
print(data4.shape)
print(data5.shape)
print(data6.shape)
```

## Creation of Master Table

```python
# Step 1: Perform Left Join between data1 (Demand Data) and data2 (Inventory Data)
master_table = pd.merge(data1, data2, on=['Product_ID', 'Warehouse_ID'], how='left')

# Step 2: Perform Left Join with data3 (Product Data)
master_table = pd.merge(master_table, data3, on='Product_ID', how='left')

# Step 3: Perform Left Join with data4 (Supplier Data)
master_table = pd.merge(master_table, data4, on='Supplier_ID', how='left')

# Step 4: Perform Left Join with data5 (Warehouse Cost Data)
master_table = pd.merge(master_table, data5, on=['Product_ID', 'Warehouse_ID'], how='left')

# Step 5: Perform Left Join with data6 (Warehouse Info Data)
master_table = pd.merge(master_table, data6, on='Warehouse_ID', how='left')

# Check the final merged dataframe
print(master_table.head())
```
